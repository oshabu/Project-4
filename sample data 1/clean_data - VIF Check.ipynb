### Data preparation

```python
random_state = 42
scoring_metric = 'recall'

# target_result:'Confirmed','Deaths','Recovered','Active'
target_result = data_train['Obesity']

# total_affecting_conditions: 'Animal fats', 'Animal Products', 'Aquatic Products, Other','Cereals - Excluding Beer',
#                             'Eggs', 'Fish, Seafood','Fruits - Excluding Wine', 'Meat', 'Milk - Excluding Butter',
#                             'Miscellaneous', 'Offals', 'Oilcrops', 'Pulses', 'Spices','Starchy Roots', 'Stimulants',
#                             'Sugar & Sweeteners', 'Sugar Crops','Treenuts', 'Vegetable Oils', 'Vegetables',
#                             'Vegetal Products','Obesity', 'Undernourished', 'Population', 'Confirmed', 'Deaths',
#                             'Recovered', 'Active'
affecting_conditions = data_train[['Animal fats', 'Animal Products', 'Aquatic Products, Other','Cereals - Excluding Beer',
                            'Eggs', 'Fish, Seafood','Fruits - Excluding Wine', 'Meat',
                            'Miscellaneous', 'Oilcrops', 'Pulses', 'Spices','Starchy Roots', 'Stimulants',
                            'Sugar Crops','Treenuts', 'Vegetables',
                            'Vegetal Products', 'Undernourished', 'Population', 'Confirmed', 'Deaths',
                            'Recovered', 'Active']]
```

```python
X=affecting_conditions.values
y=target_result.values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = random_state)
```

### Lasso model

```python
lm_lasso=Lasso(alpha = 1, max_iter = 100, tol = 1e-3)
lm_lasso.fit(X_train,y_train)
y_pred_lm_lasso = lm_lasso.predict(X_test)
mse_lm_lasso = mean_squared_error(y_test, y_pred_lm_lasso)
print("RMSE using LASSO: ", np.sqrt(mse_lm_lasso))
```

# Random Forest model

```python
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
mse_rf = mean_squared_error(y_test, y_pred_rf)
print("RMSE using RF: ", np.sqrt(mse_rf))
print("RF R2 score :", r2_score(y_test,y_pred_rf))
```

### XGBoost

```python
xgb = XGBRegressor() 
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test) 
mse_xgb = mean_squared_error(y_pred_xgb, y_test)
print("RMSE:" , np.sqrt(mse_xgb))
print("XGBoost R2 score :", r2_score(y_test,y_pred_xgb))
```
